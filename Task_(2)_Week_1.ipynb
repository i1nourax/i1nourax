## 1. Exploratory Data Analysis (EDA)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df2 = pd.read_csv('/content/Task (2) Dataset.csv')
df2
```

### Perform summary statistics on the dataset.

```python
summary_statistics = df2.describe()
print(summary_statistics)
```

### Identify and analyze patterns in the data.

```python
plt.figure(figsize=(10, 6))
sns.histplot(df2['TrafficIndexLive'], kde=True)
plt.title('Distribution of TrafficIndexLive')
plt.xlabel('TrafficIndexLive')
plt.ylabel('Frequency')
plt.show()
```

```python
city_traffic = df2.groupby('City')['TrafficIndexLive'].mean().sort_values(ascending=False)
print(city_traffic)
```

```python
df2['Datetime'] = pd.to_datetime(df2['Datetime'])
df2.set_index('Datetime', inplace=True)
plt.figure(figsize=(12, 6))
df2['TrafficIndexLive'].plot()
plt.title('Trend of TrafficIndexLive over Time')
plt.xlabel('Datetime')
plt.ylabel('TrafficIndexLive')
plt.show()
```

### Visualize the distribution of key variables (e.g., Traffic_Index, Date).

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='City', y='TrafficIndexLive', data=df2)
plt.title('Average Traffic Index by City')
plt.xlabel('City')
plt.ylabel('Average Traffic Index')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='TrafficIndexLive', y='TrafficIndexWeekAgo', data=df2)
plt.title('Relationship between TrafficIndexLive and TrafficIndexWeekAgo')
```
